```python
!pip install pandas sqlalchemy pymysql

import pandas as pd
from sqlalchemy import create_engine

# 1. Extraction : Lecture d'un fichier CSV
data = pd.read_csv('/home/tegongue/Documents/formations/Master 2 data engineering/Etl/ventes.csv')

# 2. Transformation : Nettoyage et ajout d'une colonne
data = data.dropna()  # Corriger l'oubli d'un saut de ligne
data['profit_margin'] = (data['revenue'] - data['cost']) / data['revenue']

# 3. Connexion à MySQL (docker run)
engine = create_engine('mysql+pymysql://user:admin@localhost:3306/datawarehouse')

# 4. Chargement des données dans MySQL
data.to_sql('sales_etl', engine, if_exists='replace', index=False)

print("Pipeline ETL terminé avec succès.")
print(data.head())  # Affiche les premières lignes du DataFrame
print(data.columns) 
```

```python
# Import des bibliothèques nécessaires
import pandas as pd  # Utilisé pour manipuler les données tabulaires (CSV)
from neo4j import GraphDatabase  # Utilisé pour interagir avec la base de données Neo4j

# 1. Extraction : Lecture d'un fichier CSV
file_path = '/home/tegongue/Documents/formations/Master 2 data engineering/Etl/ventes_users.csv'

# Chargement du fichier CSV dans un DataFrame pandas
df = pd.read_csv(file_path)
print("Fichier chargé avec succès.")
print(f"Colonnes du fichier : {df.columns.tolist()}")  # Liste des colonnes dans le fichier
print(f"Nombre total de lignes : {len(df)}")  # Nombre total de lignes dans le fichier

# Vérifier les colonnes pour des valeurs manquantes (NaN)
print("Validation des données : Vérification des valeurs manquantes...")
nan_counts = df.isna().sum()  # Compte le nombre de valeurs NaN par colonne
print(f"Valeurs manquantes par colonne :\n{nan_counts}")

# 2. Transformation : Nettoyage et préparation des données
print("Transformation : Remplacement des valeurs NaN...")
df['revenue'] = df['revenue'].fillna(0)  # Remplace les valeurs NaN dans la colonne 'revenue' par 0
print("Valeurs NaN remplacées.")

# Aperçu des données transformées
print("Aperçu des données transformées :")
print(df.head())  # Affiche les 5 premières lignes du DataFrame

# 3. Connexion à la base de données Neo4j
def get_neo4j_driver():
    
    #Fonction pour établir une connexion avec la base de données Neo4j.
    #Retourne un objet driver permettant d'exécuter des requêtes Cypher.
    uri = "bolt://localhost:7687"  # URL de connexion à Neo4j (à adapter si nécessaire)
    user = "neo4j"  # Nom d'utilisateur Neo4j
    password = "medarine672"  # Mot de passe Neo4j
    driver = GraphDatabase.driver(uri, auth=(user, password))  # Création du driver
    print("Connexion à Neo4j établie avec succès.")
    return driver

# 4. Chargement des données transformées dans Neo4j
def transform_and_load_to_neo4j(df):
    
    #Fonction pour transformer les données et les charger dans Neo4j.
    #Crée des nœuds (produits et utilisateurs) et des relations entre eux.
    driver = get_neo4j_driver()  # Connexion à Neo4j

    with driver.session() as session:
        # Parcours de chaque ligne du DataFrame
        for index, row in df.iterrows():
            # Récupération des données de la ligne actuelle
            product_id = row['product_id']
            product_name = row['product_name']
            revenue = row['revenue']
            cost = row['cost']
            user_name = row['user_name']
            user_age = row['user_age']
            purchase_date = row['purchase_date']
            purchase_quantity = row['purchase_quantity']

            # Affichage des informations pour suivi
            print(f"Insertion des données : Produit {product_name}, Utilisateur {user_name}...")

            # Requête Cypher pour créer les nœuds et les relations
            query = f"""
            MERGE (p:Product {{id: {product_id}, name: '{product_name}', revenue: {revenue}, cost: {cost}}})
            MERGE (u:User {{name: '{user_name}', age: {user_age}}})
            MERGE (u)-[:BOUGHT {{purchase_date: '{purchase_date}', quantity: {purchase_quantity}}}]->(p)
            """
            
            # Exécution de la requête
            try:
                session.run(query)
                print(f"Produit {product_name} et utilisateur {user_name} insérés avec succès.")
            except Exception as e:
                print(f"Erreur lors de l'insertion : {e}")

    # Fermeture de la connexion
    driver.close()
    print("Connexion à Neo4j fermée.")

# Début du processus de chargement
print("Début du chargement des données dans Neo4j...")
transform_and_load_to_neo4j(df)
print("Toutes les données ont été insérées avec succès dans Neo4j.")
```
